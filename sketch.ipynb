```
# Description: this program converts an image to a simple sketch
```

```
# Import used libraries
from google.colab import files
import cv2
# Google colab isn't able to complete the normal cv2 imshow function, so we
# must import cv2_imshow
from google.colab.patches import cv2_imshow
```

```
# Upload your chosen image to be converted to sketch. This will be uploaded to 
# the notebooks file folder
upload = files.upload()
```

```
# Giving the location of the file in the sample_data folder 
img_location = '/content/'
filename = 'huskypuppy.jpg' # or whatever name of your file
```

```
# Reading the image file
img = cv2.imread(img_location+filename)
```

```
# Showing the original file image
cv2_imshow(img)
```

```
# Convert the image to gray scale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```
# Invert the image
inverted_gray_image = 255 - gray_image
cv2_imshow(inverted_gray_image)
```

```
# Blur the imageby quassian function
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21), 0)
cv2_imshow(blurred_img)
```

```
# Invert the blurred image
inverted_blurred_img = 255 - blurred_img
cv2_imshow(inverted_blurred_img)
```

```
# Create the pencil sketch image, the scale factor will determine the pencil thickness
pencil_sketch_img = cv2.divide(gray_image, inverted_blurred_img, scale=226.0)
cv2_imshow(pencil_sketch_img)
```
